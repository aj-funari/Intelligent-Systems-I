```python
import matplotlib.pyplot as plt

from skimage import filters
from skimage import exposure

our_image = data.camera()
val = filters.threshold_otsu(our_image)

hist, bins_center = exposure.histogram(our_image)

new_image = our_image < val

plt.imshow(our_image, cmap='gray', interpolation='nearest')

plt.imshow(new_image, cmap='gray', interpolation='nearest')

````python
import matplotlib.pyplot as plt
from skimage import data
from skimage import filters
from skimage import exposure

our_image = data.camera()
val = filters.threshold_otsu(our_image)

hist, bins_center = exposure.histogram(our_image)

new_image = our_image < val

plt.imshow(our_image, cmap='gray', interpolation='nearest')
plt.show()
plt.imshow(new_image, cmap='gray', interpolation='nearest')
plt.show()
plt.plot(bins_center, hist, lw=2)
plt.axvline(val, color='k', ls='--')
plt.show()
````
